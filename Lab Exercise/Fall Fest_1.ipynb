```python
from qiskit import *
import qiskit.quantum_info as qi
qc = QuantumCircuit(2)
qc.h(1)
qc.h(0)
qc.draw()
```

```python
v = qi.Statevector(qc)
v
```

```python
d = qi.DensityMatrix(qc)
d
```

```python
probs = v.probabilities()
probs
```

```python
counts = v.sample_counts(shots=1024)
counts 
```

```python
from qiskit.visualization import plot_histogram, array_to_latex

plot_histogram(counts)
```

```python
samples = v.sample_memory(shots=2)
samples
```

```python
o = qi.SparsePauliOp(['ZZ','XX', 'II'], [0.2,-0.3,0.4])
array_to_latex(o.to_operator())
```

```python
o_exp = v.expectation_value(o)
o_exp
```
